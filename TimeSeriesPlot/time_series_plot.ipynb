```python
# Load the CSV file into a DataFrame
csv_file_path = '/Users/kelcigriffin/Desktop/Group3_Project3/clean_data/crime_rates.csv'
data = pd.read_csv(csv_file_path)
```

```python
# Select columns with "_rate" in the name
rate_columns = [col for col in data.columns if '_rate' in col]
```

```python
# Subset the DataFrame with selected columns
data_subset = data[['state_abbr', 'data_year'] + rate_columns]

data_subset
```

```python
# Convert the DataFrame to JSON and save it to a file
json_file_path = '/Users/kelcigriffin/Desktop/Group3_Project3/Visualizations/Data/filtered_crime_rates.json'
data_subset.to_json(json_file_path, orient='records', lines=True)

print(f'CSV file "{csv_file_path}" converted to JSON file "{json_file_path}" successfully.')
```

UNEMPLOYMENT DATA

```python
# Load the unemployment CSV file into a DataFrame
file_path_2 = '/Users/kelcigriffin/Desktop/Group3_Project3/clean_data/unemployment_year_state.csv'
unemployment_data = pd.read_csv(file_path_2)

unemployment_data
```

```python
# Reshape the unemployment data to a long format suitable for merging
unemployment_data_long = pd.melt(unemployment_data, id_vars=['State'], var_name='Year', value_name='Unemployment_Rate')

# Convert 'Year' to numeric in the unemployment data for proper merging
unemployment_data_long['Year'] = pd.to_numeric(unemployment_data_long['Year'])

# Create a dictionary to map full state names to abbreviations from the crime data
state_name_to_abbr = {
    state: abbr for abbr, state in zip(data_subset['state_abbr'].unique(), unemployment_data['State'].unique())
}

# Map state names to abbreviations in the unemployment data
unemployment_data_long['State_Abbr'] = unemployment_data_long['State'].map(state_name_to_abbr)

unemployment_data_long
```

```python
# Now, merge the unemployment data with the crime data
merged_data_df = pd.merge(data_subset, unemployment_data_long, left_on=['state_abbr', 'data_year'], right_on=['State_Abbr', 'Year'], how='left')

# Dropping redundant columns ('State', 'Year', 'State_Abbr') after merge
merged_data_df.drop(['State', 'Year', 'State_Abbr'], axis=1, inplace=True)


# Display the first few rows of the merged DataFrame to verify the merge
merged_data_df.head(50)
```

```python
# Convert the DataFrame to JSON and save it to a file
merged_data_json_file_path = '/Users/kelcigriffin/Desktop/Group3_Project3/Visualizations/Data/merged_unemployment_and_crime_rates.json'
merged_data_df.to_json(merged_data_json_file_path, orient='records', lines=True)

print(f'CSV file "{file_path_2}" converted to JSON file "{merged_data_json_file_path}" successfully.')
```
