```python
state_abbreviations = [
    'AL', 'AK', 'AZ', 'AR', 'CA',
    'CO', 'CT', 'DE', 'FL', 'GA',
    'HI', 'ID', 'IL', 'IN', 'IA',
    'KS', 'KY', 'LA', 'ME', 'MD',
    'MA', 'MI', 'MN', 'MS', 'MO',
    'MT', 'NE', 'NV', 'NH', 'NJ',
    'NM', 'NY', 'NC', 'ND', 'OH',
    'OK', 'OR', 'PA', 'RI', 'SC',
    'SD', 'TN', 'TX', 'UT', 'VT',
    'VA', 'WA', 'WV', 'WI', 'WY'
]
```

```python
national_crime_by_year = pd.DataFrame()

for i in state_abbreviations:
    agency_url = f"https://api.usa.gov/crime/fbi/cde/arrest/state/{i}/all?from=2004&to=2024&API_KEY={API_key}"
    query = requests.get(agency_url).json()
    temp_df = pd.DataFrame(query["data"])
    temp_df['state_abbr'] = i
    cols = ['state_abbr'] + [col for col in temp_df.columns if col != 'state_abbr']
    temp_df = temp_df[cols]
    national_crime_by_year = pd.concat([national_crime_by_year, temp_df], ignore_index=True)

national_crime_by_year
```
